```python
import os
import shutil
import pandas as pd

# Define paths

source_base_path = r'C:/Users/bengi/Downloads/LERA Dataset Ankle'
destination_base_path = r'C:/Users/bengi/Downloads/LERA Dataset Ankle Healthy'
csv_file_path = r'filtered_ankle_healthy_file.csv'

# Create the destination base path if it does not exist
#os.makedirs(destination_base_path, exist_ok=True)

# Read the CSV file
try:
    df = pd.read_csv(csv_file_path)
    print("CSV file loaded successfully.")
except FileNotFoundError:
    print("Error: CSV file not found. Please check the file path.")
    exit()
except pd.errors.EmptyDataError:
    print("Error: The CSV file is empty.")
    exit()
except pd.errors.ParserError:
    print("Error: The CSV file could not be parsed.")
    exit()

# Assuming the CSV has a single column with folder names
folder_names = df.iloc[:, 0].astype(str).tolist()  # Convert to list of strings

# Prepend base path to each folder name
folder_paths = [os.path.join(source_base_path, folder_name) for folder_name in folder_names]

# Debug: Print folder paths to check
print("Folder paths:")
print(folder_paths)

# Copy folders
for folder_name in folder_names:
    source_folder_path = os.path.join(source_base_path, folder_name)
    destination_folder_path = os.path.join(destination_base_path, folder_name)
    
    if os.path.exists(source_folder_path):
        try:
            shutil.copytree(source_folder_path, destination_folder_path)
            print(f"Created copy of {folder_name} in {destination_base_path}")
        except Exception as e:
            print(f"Error creating copy of {folder_name}: {e}")
    else:
        # Create an empty folder if the source folder doesn't exist
        try:
            os.makedirs(destination_folder_path)
            print(f"Created new folder {folder_name} in {destination_base_path}")
        except Exception as e:
            print(f"Error creating folder {folder_name}: {e}")

print("Folder copying process completed.")
```
